## Natural disasters list

The Small Business Administration has a list of natural disasters that cause extensive economic damage. Run the next cell to pull the list into memory

````python
disasters = [
        "Landslides",
        "Earthquake",
        "Wildfires",
        "Tornadoes",
        "Hurricane",
        "Wind Storm",
        "Winter Freeze",
        "Ice Storm",
        "Drought",
        "Tsunami",
        "Flash floods",
        "Mudslides",
        "Torrential rainfall",
        "Riverine floods",
        "Tidal Surge"
         ]
````

````python
## call the disaster list here
disasters
````

````python
## the first item in the list
disasters [0]
````

````python
## call the penultimate item (without manually counting the number of items)
disasters [-2]
````

### Clarification:

For the next two questions, I am referring to numerical positions rather than zero-indexed positions. In other words, you will have to translate the numerical position into zero-indexed positions in your slicing.

````python
## create a new list called disasters_subset that holds the 3rd to 6th items inclusive
disasters_subset = disasters [2:6]
disasters_subset
````

````python
## create a new list called disasters_subset2 that holds the 2nd through 8th items
disasters_subset2 = disasters [1:8]
disasters_subset2 
````

You realize that the disasters from "Tsunami" to the end of the list are forms of water damage.

Programmatically create a new list called ```water_incidents``` that includes all the disasters from "Tsunami" to "Tidal Surge".

- Do NOT manually count to figure out the index position of "Tsunami".
- Do NOT manually type out a new list.

````python
#identifico la posición de Tsunami
disasters.index("Tsunami") 
````

````python
water_incidents = disasters [disasters.index("Tsunami"):]
water_incidents
````

## Accessing data in dictionaries

The following ```dictionary``` holds all congressional committees.

````python
## run this cell
ag_com = {'committee_name': 'Agriculture',
          'branch': 'house',
          'chair': 'GLENN "GT" THOMPSON',     
  'link': 'https://www.congress.gov/committee/house-agriculture/hsag00'}
 
````

````python
ag_com
````

````python
## retrieve the committee name
ag_com.get("committee_name")
````

````python
## retrieve the link for more info
ag_com.get("link")
````

## Combine the 4 dataframes


````python
## create a list to combine all 
````

````python
ems_list = [ems_19_first_df, ems_19_last_df,ems_20_first_df,ems_19_last_df]
````
