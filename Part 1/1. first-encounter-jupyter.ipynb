## Variables

Variables are placeholders for values. They can be any of the following:

-   **Numbers**
-   **Strings**, that are made up of letter, words, sentences, etc.
-   **Placeholders**, that receive their value from inputs, computations or concatenation of other variables.
-   **Lists**, that holds a multiple values in one variable.
-   **Dictionaries**, variables that hold interrelated data or info as key-value pairs)
-   **Functions**, bits of code that you can reuse...(more on this at some point this semester!).

...and many many other types of data.



### Naming Variables

The idea is to give meaningful names so you aren't confused by what they stand for. Apart from that, here are some basic rules:

-   They ARE case sensitive.
-   Cannot begin with numbers.
-   Make them unique (just like IDs, you should only have ONE variable with a particular name).
-   No spaces between words, but don't use compound words (```monthlyexpensessandeep```) because they are hard to read. Instead use:
    - camelCase (```monthlyExpensesSandeep```)
    - **underscores or snake_case: (```monthly_expenses_sandeep```)**

    
-   Don't use any Python programming terms [reserved words](https://betterprogramming.pub/all-33-reserved-keywords-in-python-98e55f6c8ded).
-   Think of code as story. Each variable is like a character/actor. Naming each in a relevant way such makes the narrative (the code) easier to follow.
-   Keep the variables short but make sure they make sense.


````python
## Calculate total monthly rent here based on variables
m_rent_alfred = 1_000
m_rent_maggie = 750
m_rent_fatima = 1_200
m_rent_total = m_rent_alfred + m_rent_maggie + m_rent_fatima
````

````python
## What is Maggie's share of the total rent?
s_maggie = (m_rent_maggie / m_rent_total) * 100
s_maggie
````

# f-string literals

## f-string literals to clarify
```print(f"Some identifier words: {actual_variable}")```


````python
## print total all the monthly rents plus totals using f-strings
print (f"Maggie's rent is ${m_rent_maggie:,}")
print (f"Total montly rent is ${m_rent_total:,}")
print(f"Alfred's anual rent is ${m_rent_alfred * 12:,}") #separadores de comas
````

## f-string literals to prettify



````python
## control decimal spaces
print(f"Maggie's share of the rent is {s_maggie:.0F}%") #.2F marca los dos espacios decimales
````

````python
## let's try on this big number
big_number = 2_000_163_451.8306490456
````

````python
## print big_big number to one decimal place, with thousands separator.
print(f"{big_number:,.1F}")
````

## Dictionaries

A Python Dictionary is another common and powerful datatype.

Think of it as a collection of data points about a single element.

For example, a car model:

- Brand: Honda
- Type: SUV
- Model: CRV
- Trim: Leather
- cost: 40_000
- on_sale: True
- VIN: 2973

A dictionary would hold this data as **key: value** pairs

A dictionary is wrapped in ```{}```.

Each key: value pair is separated by a comma.

The format is ```{"key1_string": "string_value", "key2_number": 12, "key3_bolean": True}```

````python
## build the car_1 dictionary
carr_1 = {
    "Brand": "Honda",
    "Type": "SUV", 
    "Model": "CRV", 
    "Trim": "Leather", 
    "cost": 40_000, 
    "on_sale": True, 
    "VIN": 2973
}
````

````python
## call car_1
carr_1
````

````python
## what car_1's datatype
type(carr_1)
````

Build me a dictionary called car_2 with the following data points.

Note: In this case the vin is first. The order in a dictionary doesn't matter.

- vin: 8303
- Brand: Honda
- Type: Sedan
- Model: CRV
- Trim: Cloth
- cost: 35_000
- on sale: no


````python
## build car_2
car_2 = {
    "VIN": 8303, 
    "Brand": "Honda",
    "Type": "Sedan", 
    "Model": "CRV", 
    "Trim": "Cloth", 
    "cost": 35_000, 
    "on_sale": False,
}
````

````python
## call car_2
car_2
````
