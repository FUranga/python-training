## Read two or more separate files and combine them.

Often, we have to work with ```bike_data_2019.csv```, ```bike_data_2020.csv```, etc.

But if we really want to see change over time, we need to combine these spreadsheets into a single spreadsheet.

Bring in ```flt_delays_17.csv``` and ```flt_delays_18.csv``` and combine them into a dataframe called ```delay_df```

````python
df_2017 = pd.read_csv("data/flt-delays-17.csv")
df_2017.head(2)
````

````python
df_2018 = pd.read_csv("data/flt-delays-18.csv")
df_2018.head(2)
````

````python
## confirm they are in similar columns
df_2017.info()
````

````python
## confirm they are in similar columns
df_2018.info()
````

````python
## If we only want to see how many rows and columns
df_2017.shape
````

````python
df_2018.shape
````

````python
## confirm they are in similar columns
df_2018.shape[0]+df_2017.shape[0]
````

````python
## concat the two by placing both dataframes in a list
df_delays = pd.concat([df_2017,df_2018])
df_delays.shape
````

````python
## How many total rows?
df_delays.shape[0]
````

````python
## what type of data is a shape
type(df_delays.shape)
````

````python
## we can pull out values from a tuple, just like a list or dictionary
````

````python
## the number of rows should add up correctly
## Dos == no son como hacer una asignacion de valor si no que pregunta si es igual, valida
df_2018.shape[0] + df_2017.shape[0] == df_delays.shape[0]
````

````python
### Get a sample of the new df
df_delays.sample (10)
````

# Retrieving data

We can store data in variables, lists and dictionaries. We'll also want to pull data from these objects.

## retrieving data from lists...

### 1. List basics:

````python
## Run this cell to pull this list into memory:

toxic_chemicals = [
    "Polyvinyl Chloride",
    "PVC",
    "Perfluorooctanoic Acid",
    "PFOA",
    "Triclosan",
    "Bisphenol-A",
    "BPA",
    "Formaldehyde",
    "Naphthalene",
    "Asbestos"
]
````

````python
## call the list
toxic_chemicals
````

````python
## print the list
print (toxic_chemicals)
````

````python
## how long is this list?
len(toxic_chemicals)
````

````python
#Desde el comienzo hasta el numero 4 (inclusive, o el 5 excluido)
toxic_chemicals[:5]
````

````python
#Desde el cuatro hasta el final
toxic_chemicals[4:]
````

### 2. Slicing to target items in a list

````python
## Last time we used index() method to retrieve 
## the index position of an item:
toxic_chemicals.index("BPA")
````

#### Recall that the first value in almost every programming language is zero-indexed, or the count always starts at zero.

#### Either way, what we really want is to retrieve the data, not the index position

#### Now we want to retrieve the data, not the index

```list_name[index_position]```

````python
toxic_chemicals [6] #fifth value
````

````python
## call the penultimate item in the list
toxic_chemicals[-2]
````

#### How about retrieving a subset of data and not just individual data points?

#### Slice notation

````python
## Call the first 3 items:
toxic_chemicals[0:3]
````

## Retrieving data from a ```dictionary```

```dictionary_name.get("key")```

````python
## run this single dictionary
usa = {'GDP': '$20.4 trillion',
 'NATO': True,
 'capital': 'Washington, DC',
 'country': 'United States',
 'currency': 'Dollar',
 'population': 329450000}
````

````python
## is the USA a member of nato?
usa.get('NATO')
````

````python
## what is the US currency?
usa.get('currency')
````

## Retrieving data from a list of dictionaries

````python
## call our countries dictionary just to see it again
country_stats
````

````python
## Is the 2nd dictionary item a member of NATO?
country_stats[1].get('NATO')
````

````python
## what is capital of the 3rd dictionary item?
country_stats[2].get('capital')
````
