## Additional steps
### Re-import the database
### Remove commas from the 'Median household income column'
### Convert to numeric

```python
df = pd.read_csv("data/cable-bill-only.csv")
df.head()
```

```python
#Remove ,
df ['Median household income'] = df['Median household income'].str.replace(',','')
df
```

```python
#Convert to float
df['Median household income'] = pd.to_numeric(df['Median household income'])
df.info()
```

```python
### Calculate percentage change with respect to 1995: (new data - 1995 data) / 1995 data * 100 
df["bill_pct_change"] = (df["Avg. cable bill"] / df["Avg. cable bill"].iat[0]  - 1) * 100
df["income_pct_change"] = (df["Median household income"] / df["Median household income"].iat[0]  - 1) * 100
df ["cola_pct_change"] = df ["Social Security cost-of-living adjustment"] - df ["Social Security cost-of-living adjustment"] .iat[0]
df
```

```python
df.columns
```

```python
# Create a graphic
df.plot(x="Year",
       y=['bill_pct_change','income_pct_change', 'cola_pct_change'],
       title = "Cable bill v. income and cost of living (Percentage change compared to 1995)",
       style = {"bill_pct_change":"red",
               "income_pct_change":"blue",
               "cola_change":"gray"},)
```

### The main findings are:

#### 1. The cost of living and the average household income increased at a similar rate during the period. In fact, incomes grew faster. Median income increased by 120% between 1995 and 2022, while living costs rose by 73%.

#### 2. Cable bills have always increased faster than the cost of living and income, but there was a turning point in 2016. Between 1993 and 2015, the cumulative increase was 250%. That is, four times higher than income and cost of living in the same period. But in 2016 it skyrocketed: it more than doubled that year and continued to increase at a high rate until accumulating in 2022 an increase of more than 1000% over 1995.
